```python
    """
    The role of this file is to provide the functions used to 
    transmit the parsed data to the SPI.

    Some of the following things might have to be removed when going for the SPI :
    - time library 
    - len_last_wr_data

    TO CHANGE : FOR TEST PURPOSES WE RECORD THE WRITTEN DATA : IN FACT, WE GET THE DATA BACK FROM WRITTING AND NOT LOOKING AT THE MEM ...

    """
    
    # Tests if the given str argument is null or empty
    # This function should be moved 
    def is_empty(arg : str):
        if (arg is None or arg == ""):
            return True 

    # command_disjunction only gives the first level of behaviour of each command
    # returns the list of rd_data to executue the compare command --> might be changed ...
    def command_disjunction(AxiQspi, result_file, parsed_line : list[str], rd_data_queue : list[bytearray]) -> None :
        command = parsed_line[0]
        dict_spi_data_len = transmitted_data_len_dict # from package.ipynb
        if command in dict_spi_data_len :
            data_len = dict_spi_data_len[command]

        if ((command == "REGRD" )):
            spi_cmd = spi_commands["SPI_REGRD"]
            rd_addr = parsed_line[1]
            last_rd_data = rd_transaction_behaviour(AxiQspi, result_file, command, spi_cmd, rd_addr, data_len)
            # rd_data_queue.append(last_rd_data)

        elif ((command == "REGWR" )):
            spi_cmd = spi_commands["SPI_REGWR"]
            wr_addr = parsed_line[1]
            wr_data = parsed_line[2::]
            #TO CHANGE 
            last_wr_data = wr_transaction_behaviour(AxiQspi, result_file, command, spi_cmd, wr_addr, wr_data)
            rd_data_queue.append(last_wr_data)

        elif (command == "MEMRD") :
            spi_cmd = spi_commands["SPI_MEMRD"]
            rd_addr = parsed_line[1]
            last_rd_data = rd_transaction_behaviour(AxiQspi, result_file, command, spi_cmd, rd_addr, data_len)
            # rd_data_queue.append(last_rd_data)

        elif (command == "MEMWR"):
            spi_cmd = spi_commands["SPI_MEMWR"]
            wr_addr = parsed_line[1]
            wr_data = parsed_line[2::]
            #TO CHANGE 
            last_wr_data = wr_transaction_behaviour(AxiQspi, result_file, command, spi_cmd, wr_addr, wr_data)
            rd_data_queue.append(last_wr_data)

        elif (command == "MEMRDM") :
            spi_cmd = spi_commands["SPI_MEMRDM"]
            rd_addr = None
            last_rd_data = rd_transaction_behaviour(AxiQspi, result_file, command, spi_cmd, rd_addr, data_len)
            # rd_data_queue.append(last_rd_data)

        elif (command == "MEMWRM") :
            spi_cmd = spi_commands["SPI_MEMWRM"]
            wr_addr = None
            wr_data = parsed_line[1::]
            #TO CHANGE 
            last_wr_data = wr_transaction_behaviour(AxiQspi, result_file, command, spi_cmd, wr_addr, wr_data)
            rd_data_queue.append(last_wr_data)

        elif ((command == "REGCMP")|(command == "MEMCMP")):
            data_ref = parsed_line[1::] 
            if (is_empty(rd_data_queue)):
                no_rd_data(result_file)
            else :
                last_rd_data = rd_data_queue.pop()
                # print(f" data_ref is {data_ref} and last_rd_data is {last_rd_data}")
            cmp_behaviour(result_file, command, data_ref, last_rd_data)

        else :
            unknown_cmd(result_file, command)
        return rd_data_queue




    # Implements the basic behaviour of the read commands
    def rd_transaction_behaviour (AxiQspi, result_file, command : str, spi_cmd : str, rd_addr : str, data_len : int) -> bytearray :
        rd_line = rd_spi_transaction(AxiQspi, spi_cmd, rd_addr, data_len)
        formated_cmd, formated_addr, formated_data = format_data(rd_line)
        # result_file.write(rd_msg(formated_cmd, formated_addr, formated_data))
        # print("last rd data :", formated_data)
        return formated_data



    # Implements the basic behaviour of the write commands
    def wr_transaction_behaviour(AxiQspi, result_file, command : str, spi_cmd :str, wr_addr : str, wr_data : list[str]) -> str:
        written_bytes = wr_spi_transaction(AxiQspi,spi_cmd, wr_addr, wr_data)
        formated_cmd, formated_addr, formated_data = format_data(written_bytes)
        # display the data to be written
        wr_msg(result_file, command, wr_addr, wr_data)
        # display the written data
        wr_sent_back(result_file, formated_cmd, formated_addr, formated_data)
        
        # TO CHANGE
        return formated_data

    
    # Implements the basic behaviour of the cmp commands
    # data_ref is composed of a certain nb of 8 hex characteres packets : its the parsed line format
    # elements of rd_data_queue have the following format : 
    # [0xcommand_byte, 0xaddr_byte(s), [0xdata_byte[0], ... , 0xdata_byte[i] , ...]]
    def cmp_behaviour(result_file, command : str, data_ref : list[str], last_rd_data : list[str]) -> None :
        starting_cmp(result_file)
        nb_rd_bytes = len(last_rd_data)
        len_ref_data = len(data_ref)
        result = 0
        if (len_ref_data != nb_rd_bytes): 
            # print("these are the number of bytes mismatch :", len_ref_data, nb_rd_bytes)
            number_of_packets_mismatch(result_file, len_ref_data, nb_rd_bytes)
            result += 1
        else :
            #Here from the MSB to the LSB 
            for hexa_packet_cnt in range(len_ref_data):
                rd_hex_packet = last_rd_data[hexa_packet_cnt]
                hex_packet_ref = data_ref[hexa_packet_cnt]
                if (command == "REGCMP") :
                    storage = "reg"
                elif (command == "MEMCMP") :
                    storage = "mem"
                result += equal_bytes(result_file, hex_packet_ref, rd_hex_packet, storage) # if not equal : raise Exception : see messsages.ipynb
        if (result == 0):    
            valid_cmp(result_file, data_ref, last_rd_data)
        else :
            invalid_cmp(result_file, data_ref, last_rd_data)



    # Take the 8 hex values and transforms it into list of 2 hex packets
    def parse_packet (data : str) -> list[str] :
        bytes_list = []
        if (len(data)%2 == 1):
            data = '0' + data  
        for i in range(0, len(data), 2) :
            hex_str = '0x' + data[i:i+2]
            bytes_list.append(int(hex_str, 16))
        return bytes_list

    # 
    def format_data (data_line : list[list[int]]) :
        spi_cmd = data_line[0][0]
        
        # Formated command :
        formated_cmd = from_int_to_hex(spi_cmd)
        
        # Formated address :
        # MEMWRM command : no address
        if (len(data_line)<3):  
            data = data_line[1]
            formated_addr = None
        
        # REGWR or MEMWR command
        else : 
            addr = data_line[1]
            data = data_line[2]
            formated_addr = ''
            for addr_byte in addr :
                formated_addr += from_int_to_hex(addr_byte)
            
        # Formated data :
        # We try to build fornated_data the same way data is extracted from the test file
        formated_data = from_list_int_to_hex(data)
            
        return formated_cmd, formated_addr, formated_data

    # Test if two given bytes are equal and writes the result in the result file 
    # TO CHANGE TO COMPARE BITs !
    def equal_bytes(result_file, hex_packet_ref : str, rd_hex_packet : str, storage : str) -> int :
        if (hex_packet_ref.upper() != rd_hex_packet.upper()): # in case hex data differs
            cmp_mismatch(result_file, hex_packet_ref, rd_hex_packet, storage)
            return 1
        else :
            cmp_match(result_file, hex_packet_ref, rd_hex_packet, storage)
            return 0


    # get_bytes returns the list of bytes stored at a given address (addr)
    def rd_spi_transaction(AxiQspi, spi_cmd : str, addr : str, data_len : int) -> bytearray:
        rd_line = []

        # Send command to SPI
        xfer(AxiQspi, [spi_cmd])


        # Send address to SPI
        # MEMWR or REGWR command 
        if (addr): 
            addr_hex_list = parse_packet(addr)
            
            # receive cmd
            rd_line.append(xfer(AxiQspi, [addr_hex_list[0]]))
        
            # receive addr
            if (len(addr_hex_list)>1):
                data_to_send = [0x00, 0x00]
            else :
                data_to_send = [0x00]
            rd_line.append(xfer(AxiQspi, data_to_send))
            
        
        # Receiving the read values 
        for nb_data_in_the_line in range(data_len + 1): # we add one to get the last data byte 
            rd_line.append(xfer(AxiQspi, [0x00])) 
        return rd_line 



    # write_val writes the list of bytes (data) at a given address (addr)
    # the format of written_line is : [cmd_byte,[addr_bytes],[data_bytes]]
    def wr_spi_transaction(AxiQspi, spi_cmd : str, addr : str, data : list[str]) -> None :
        written_line = []
        data_bytes = []
        first_data_packet = parse_packet(data[0])
        
        # Send command to SPI
        xfer(AxiQspi, [spi_cmd]) # we don't get the first byte


        # Send address to SPI
        if (addr):
            addr_hex_list = parse_packet(addr)
          
            # receive cmd 
            written_line.append(xfer(AxiQspi, [addr_hex_list[0]]))

            # receive addr
            if (len(addr_hex_list)>1):
                data_to_send = [addr_hex_list[1], first_data_packet[0]]
            else :
                data_to_send = [first_data_packet[0]]
            written_line.append(xfer(AxiQspi, data_to_send))
            
        else :
            # receive cmd 
            data_to_send = [first_data_packet[0]]
            written_line.append(xfer(AxiQspi, data_to_send))
        
        
        # receive data 
        data_bytes = xfer(AxiQspi, first_data_packet[1::])
        for hex_packet in data[1::] : #Here hex_packet is an 8 hex values
            formated_packet = parse_packet(hex_packet)
            data_bytes += xfer(AxiQspi, formated_packet)
            
        data_bytes += xfer(AxiQspi, [0x00])
        written_line.append(data_bytes)
        return written_line 
```
