```python
# For HTML Result files 
##### Uncomment this to change the default color of msg

# Colors : can be changed in config file
text_colors = config_data["RESULT_FILE"]["html_config"]["text_colors"]
highlight_colors = config_data["RESULT_FILE"]["html_config"]["highlight_colors"]
start_p_html = f"<p style=\"color:{text_colors['default_color']};\">"
start_error_span = f"<span style=\"color:{text_colors['error_color']};  background-color: {highlight_colors['error_color']};\">"
start_valid_span = f"<span style=\"color:{text_colors['validation_color']}; background-color: {highlight_colors['validation_color']};\">"

end_cmp_div = "</div> \n"
end_span = "</span>"
end_p = "</p>"

error_msg_html = f"{start_error_span} ERROR : {end_span}"
valid_msg_html = f"{start_valid_span} SUCCESSFUL : {end_span}"

# This function generates a string containing the html content of result files :
def initiate_html_content(title : str, style_css : str, main_header_content : str) -> str :
    content = f"<!DOCTYPE html> \n "
    content += f"<html>\n<head> \n \
    <title>{title}</title> \n \
    <style>  {style_css} </style> \n \
    </head> \n <body> \n \
    <div class = \"main\"> \n "

    content += f"<h1> {main_header_content} </h1> \n"
    return content 

# write content inside a p tag in html
def write_html_msg(result_file_html, content : str) -> None :
    if (content == "" or content is None) :
        content = "No content given : "
        html_msg = start_p_html + content + end_p + "\n"
    else :
        html_msg = start_p_html + content + end_p + "\n"
    result_file_html.write(html_msg) 
    
def start_cmp_div() -> str :
    content = f" <div class= \"compare\"> \n "
    content += f" <h2 class= \"compare\"> Comparison of read data and reference data : </h2> \n "
    return content 

def end_html() -> str :
    return "</div> \n </body> \n </html>"
    

# For print if I want to implement some 
red_color = '\033[7;49;91m'
green_colour = '\033[7;49;32m'
normal_colour = '\033[0;5;49m'

error_msg = f"{red_color} ERROR : {normal_colour}"
valid_msg = f"{green_colour} SUCCESSFUL : {normal_colour}"
```

These messages are error messages :

```python
def test_config_error() -> None :
    message = f""" {error_msg} Concurrency in the launch config : see the yaml file :  
                    single test and all_tests booleans should not have the same value ! \n"""
    raise Exception(message)
    
def number_of_packets_mismatch(result_file_html, result_file_txt, nb_test_bytes : int, nb_rd_bytes : int) -> None :
    message = f" {error_msg} Both number of hex packets does not match : got {nb_rd_bytes}; should be {nb_test_bytes} \n"
    if (activate_msg_prints):
        print(message)
    if (activate_result_file_txt):
        message_txt = f"ERROR : Both number of hex packets does not match : got {nb_rd_bytes}; should be {nb_test_bytes} \n"
        result_file_txt.write(message_txt)
    if (activate_result_file_html):
        message_html = f" {error_msg_html} Both number of hex packets does not match : got {nb_rd_bytes}; should be {nb_test_bytes} \n"
        write_html_msg(result_file_html, message_html)
    if (activate_exceptions) :
        raise Exception(message)
        

def no_rd_data(result_file_html, result_file_txt) -> None :
    message = f"{error_msg} No data has been read yet. \n"
    if (activate_msg_prints):
        print(message)
    if (activate_result_file_txt):
        result_file_txt.write(message)
    if (activate_result_file_html):
        message_html = f"{error_msg_html} No data has been read yet. \n"
        write_html_msg(result_file_html, message_html)
    if (activate_exceptions) :
        raise Exception(message)

def unknown_cmd(result_file_html, result_file_txt, command : str) -> None :
    message = f" {error_msg} Unknown command: {command}"
    if (activate_msg_prints):
        print(message)
    if (activate_result_file_txt):
        message_txt = f"ERROR : Unknown command: {command}"
        result_file_txt.write(message_txt)
    if (activate_result_file_html):
        message_html = f" {error_msg_html} Unknown command: {command} \n"
        write_html_msg(result_file_html, message_html)
    if (activate_exceptions) :
        raise Exception(message)

def error_parsing_line(result_file_html, result_file_txt, line : str):
    message = f"{error_msg} unable to parse line: {line}"
    if (activate_msg_prints):
        print(message)
    if (activate_result_file_txt):
        message_txt = f"ERROR : unable to parse line: {line}"
        result_file_txt.write(message_txt)
    if (activate_result_file_html):
        message_html = f"{error_msg_html} unable to parse line: {line} \n"
        write_html_msg(result_file_html, message_html)
    if (activate_exceptions) :
        raise Exception(message)
        
        
def wrong_storage_given(result_file_html, result_file_txt, storage : str) -> None :
    message = f""" {error_msg} Wrong arg storage given : {storage}; should be either \"reg\" or \"mem\" \n"""
    if (activate_msg_prints):
        print(message)
    if (activate_result_file_txt):
        message_txt = f""" ERROR : Wrong arg storage given : {storage}; should be either \"reg\" or \"mem\" \n"""
        result_file_txt.write(message_txt)
    if (activate_result_file_html):
        message_html = f""" {error_msg_html} Wrong arg storage given : {storage}; should be either \"reg\" or \"mem\" \n"""
        write_html_msg(result_file_html, message_html)
    if (activate_exceptions) :
        raise Exception(message)
    
def cmp_mismatch(result_file_html, result_file_txt, ref_data : str, rd_data : str, storage : str) -> None :
    if (storage == "reg") :
        message = f""" **** {error_msg} \t Command #%4d: Last Reg read value (masked) = {rd_data} 
                                                       is NOT EQUAL to compared value = {ref_data} **** \n """
        message_txt = f""" **** ERROR : \t Command #%4d: Last Reg read value (masked) = {rd_data} 
                                                       is NOT EQUAL to compared value = {ref_data} **** \n """
        message_html = f""" **** {error_msg_html} \t Command #%4d: Last Reg read value (masked) = {rd_data} 
                                                                 is NOT EQUAL to compared value = {ref_data} **** \n """
        
    elif (storage == "mem") :
        message = f""" **** {error_msg} \t Command #%4d: Last Mem read value = {rd_data} 
                                              is NOT EQUAL to compared value = {ref_data} **** \n """
        message_txt = f""" **** ERROR : \t Command #%4d: Last Mem read value = {rd_data} 
                                              is NOT EQUAL to compared value = {ref_data} **** \n """
        message_html = f""" **** {error_msg_html} \t Command #%4d: Last Mem read value = {rd_data} 
                                                        is NOT EQUAL to compared value = {ref_data} **** \n """
    else :
        # message is not defined here but raises an exception in wrong_storage_given.
        wrong_storage_given(result_file_html, result_file_txt, storage)
    
    if (activate_msg_prints):
        print(message)
    if (activate_result_file_txt):
        result_file_txt.write(message_txt)
    if (activate_result_file_html):
        write_html_msg(result_file_html, message_html)
        
    if (activate_exceptions) :
        raise Exception(message)
        
def invalid_cmp(result_file_html, result_file_txt, ref_data : list[str], rd_data : list[str]) -> None :
    str_ref_data = ''
    str_rd_data = ''
    for data in ref_data :
        str_ref_data += data + '\t'
    for data in rd_data :
        str_rd_data += data + '\t'
        
    message = f" {normal_colour} \
    The compared data are not the same : {error_msg} \n \
        the ref data is : {str_ref_data} \n \
        the tested byte is : {str_rd_data} \n "
    if (activate_msg_prints):
        print(message)
    if (activate_result_file_txt):
        message_txt = f" \
        The compared data are not the same : ERROR : \n \
            the ref data is : {str_ref_data} \n \
            the tested byte is : {str_rd_data} \n "
        result_file_txt.write(message_txt)
    if (activate_result_file_html):
        message_html = f""" 
        The compared data are not the same : test {error_msg_html} <br> 
            the ref data is : {str_ref_data} <br> 
            the tested byte is : {str_rd_data} 
            """
        write_html_msg(result_file_html, message_html) 
        result_file_html.write(end_cmp_div) # end the compare div
    
    if (activate_exceptions) :
        raise Exception(message)
  
```

```python
# TO CHANGE : NOT UP TO DATE 
def rd_msg (result_file_html, result_file_txt, command : str,  rd_addr : str, rd_data : str) -> None :
    message = f"Command {command} at address {rd_addr} : Finished Reading; Read value = {rd_data} \n " 
    if (activate_msg_prints):
        print(message)
    if (activate_result_file_txt):
        result_file_txt.write(message)
    if (activate_result_file_html):
        write_html_msg(result_file_html, message)


def wr_msg (result_file_html, result_file_txt, command : str, wr_addr: str, wr_data : list[str]) -> None :
    data_joined = ""
    for data in wr_data :
        data_joined += data + '\t'
    message = f"Command {command} at address {wr_addr}: Finished Writing             {data_joined}  \n"  
    if (activate_msg_prints):
        print(message)
    if (activate_result_file_txt):
        result_file_txt.write(message)
    if (activate_result_file_html):
        write_html_msg(result_file_html, message)

# TO DELETE AFTER TESTS
def wr_sent_back(result_file_html, result_file_txt, command : str, wr_addr : str, wr_data : str) -> None :
    data_joined = ""
    for data in wr_data :
        data_joined += data + '\t'
    message = f"Command {command} at address {wr_addr}: Finished Reading from send_back {data_joined}  \n" 
    if (activate_msg_prints):
        print(message)
    if (activate_result_file_txt):
        result_file_txt.write(message)
    if (activate_result_file_html):
        write_html_msg(result_file_html, message)
    
def starting_cmp(result_file_html, result_file_txt) -> None :
    message = f"Comparison of read data and reference data \n" 
    if (activate_msg_prints):
        print(message)
    if (activate_result_file_txt):
        message_txt = f" \n Comparison of read data and reference data \n" 
        result_file_txt.write(message_txt)
    if (activate_result_file_html):
        html_cmp = start_cmp_div()
        result_file_html.write(html_cmp) 

# def cmp_mismatch_mem(test_bytes : bytearray, rd_bytes : bytearray) -> str :
#     message = f""" **** ERROR: Command #%4d: Last Mem read value = {rd_bytes}
#     is NOT EQUAL to compared value = {test_bytes} **** \n """
#     return message
```

These messages are validation messages :

```python
def cmp_match(result_file_html, result_file_txt, ref_data : str, rd_data : str, storage : str) -> None :
    if (storage == "reg"):
        message = f""" **** Command #%4d: Last Reg read value (masked) = {rd_data}
        is equal to compared value =      {ref_data} **** \n """
        message_html = f""" 
        **** Command #%4d: Last Reg read value (masked) = {rd_data}
        is equal to compared value =                      {ref_data} **** 
        """
    elif (storage == "mem"):
        message = f""" **** Command #%4d: Last Mem read value = {rd_data}
        is equal to compared value =      {ref_data} **** \n """
        message_html = f""" 
        **** Command #%4d: Last Mem read value = {rd_data}
        is equal to compared value =             {ref_data} **** 
        """   
    else :
        wrong_storage_given(result_file_html, result_file_txt, storage)
        
    if (activate_msg_prints):
        print(message)
    if (activate_result_file_txt):
        result_file_txt.write(message)
    if (activate_result_file_html):
        write_html_msg(result_file_html, message_html)
    

def valid_cmp(result_file_html, result_file_txt, ref_data : list[str], rd_data : list[str]) -> None:
    spaced_ref_data = ''
    spaced_rd_data = ''
    for data in ref_data :
        spaced_ref_data += data + '\t'
    for data in rd_data :
        spaced_rd_data += data + '\t'
        
    message = f""" {normal_colour} The compared data are the same : {valid_msg}
        the ref data is : {spaced_ref_data} 
        the read byte is : {spaced_rd_data} \n \n"""
    
    message_txt = f""" The compared data are the same : test SUCCESSFUL :
        the ref data is : {spaced_ref_data} 
        the read byte is : {spaced_rd_data} \n \n"""
        
    # Inside a <p>, use <br> to go to a new line 
    message_html = f""" 
    The compared data are the same : test {valid_msg_html} <br> 
        the ref data is : {spaced_ref_data} <br>
        the tested data is : {spaced_rd_data} 
        """
    if (activate_msg_prints):
        print(message)
    if (activate_result_file_txt):
        result_file_txt.write(message_txt)
    if (activate_result_file_html):
        write_html_msg(result_file_html, message_html)
        result_file_html.write(end_cmp_div) # end the compare div


# def cmp_match_mem(test_bytes : bytearray, rd_bytes : bytearray) -> str :
#     message = f""" **** Command #%4d: Last Mem read value = {rd_bytes}
#     is equal to compared value = {test_bytes} **** \n """
#     return message
```
